# ML model

```python
new_df # data after removing outliers
```

```python
new_ml = new_df.copy()
new_ml.head()
```

# converting to numerical data

```python
new_ml['gender'] = new_ml['gender'].map({'Male':0,'Female':1})
new_ml['usertype'] = new_ml['usertype'].map({'Subscriber':0,'Customer':1})
new_ml.head()
```

```python
from sklearn.preprocessing import LabelEncoder
ml = new_ml.copy()
enc = LabelEncoder()
ml['from_station_name'] = enc.fit_transform(ml['from_station_name'] )
ml['to_station_name'] = enc.fit_transform(ml['to_station_name'] )
ml.head()
```

```python
ml.drop('birthyear',axis=1,inplace=True)
ml.head()
```

```python
ml.head()
```

```python
ml['start_time'] = enc.fit_transform(ml['start_time'])
ml['end_time'] = enc.fit_transform(ml['end_time'])
ml['month'] = enc.fit_transform(ml['month'])

ml.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = pd.DataFrame(scaler.fit_transform(ml),columns=ml.columns)
scaled
```

```python
X = scaled.drop('tripduration',axis=1)
y = pd.DataFrame(scaled['tripduration'])
```

```python
# splitting data for training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
X_train
```

```python
X_test
```

```python
scaled.head()
```

```python
# using Linear Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
y_pred_df = pd.DataFrame(y_pred,columns=['predicted'])
y_pred_df
```

```python
pred_df = pd.DataFrame(columns=['Actual','predicted'])
pred_df['Actual'] = y_test.reset_index()['tripduration']
pred_df['predicted'] = y_pred_df.reset_index()['predicted']
pred_df['Error'] = pred_df['Actual'] - pred_df['predicted']
pred_df
```

```python
from sklearn.metrics import r2_score
```

```python
accuracy = r2_score(y_test,y_pred_df)
print("test accuracy",accuracy)
```
